### Completing our data

```python
import pandas as pd
import json

with open('../data/raw-newWc/worldcup.json', 'r', encoding='utf-8') as archivo:
    datos_torneo = json.load(archivo)

df_mundial = pd.DataFrame(datos_torneo['matches'])

df_limpio = df_mundial[['round', 'date', 'team1', 'team2']].copy()

df_limpio.rename(columns={
    'team1': 'home_team_name',
    'team2': 'away_team_name',
    'round': 'stage'
}, inplace=True)

df_limpio.to_csv('../data/processed/calendario_2026_limpio.csv', index=False)

```
